**Separating dependent variables and independent variables**

```python
x=data.drop('price_range',axis=1)
y=data['price_range']
```

```python
x.shape
```

**Standard scaling**

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x=std.fit_transform(x)
x=pd.DataFrame(x)
```

**One-Hot Encoding**

```python
y_encoded=pd.get_dummies(y)
```

**Splitting data into train and test data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y_encoded,test_size=0.2,random_state=42)
```

# **Artificial Neural Network**

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(units=10,activation='relu'))

model.add(Dense(units=4,activation='softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
y_predictions=model.predict(x_test)
```

```python
y_predictions
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

# **Hypertuning using Optuna**

```python
pip install optuna tensorflow
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def create_model(trial):
    # Define the hyperparameters to be tuned
    units = trial.suggest_int('units', 32, 512, log=True)
    dropout_rate = trial.suggest_uniform('dropout_rate', 0.2, 0.5)
    learning_rate = trial.suggest_loguniform('learning_rate', 0.001, 0.1)
    epochs = trial.suggest_int('epochs', 5, 100)

    # Build the neural network model
    model = Sequential()
    model.add(layers.Dense(units, activation='relu', input_shape=(20,)))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(4, activation='softmax'))

    # Compile the model
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model, epochs
```

```python
def objective(trial):
    # Create the model
    model, epochs= create_model(trial)

    # Train the model
    history = model.fit(x_train, y_train, epochs=epochs, validation_data=(x_test, y_test), verbose=0)

    # Evaluate the model on the validation set
    score = model.evaluate(x_test, y_test, verbose=0)

    # Return the validation accuracy as the objective value to be minimized
    return 1.0 - score[1]
```

```python
import optuna
# Create an Optuna study object and specify the optimization direction
study = optuna.create_study(direction='minimize')

# Optimize the objective function
study.optimize(objective, n_trials=50)
```

```python
trial = study.best_trial
print('Loss: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
```
