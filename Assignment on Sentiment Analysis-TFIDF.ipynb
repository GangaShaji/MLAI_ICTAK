**Step1: Removing Punctuation Marks**

```python
import string
```

```python
string.punctuation
```

```python
punctuationfree=''
def remove_punc(text):
  punctuationfree=''.join([i for i in text if i not in string.punctuation])
  return punctuationfree
```

```python
data['tweet']=data['tweet'].apply(lambda x:remove_punc(x))
```

```python
data
```

**Step2: Coverting into lower case**

```python
data['tweet']=data['tweet'].apply(lambda x:x.lower())
```

```python
data
```

**Step3: Word Tokenization**

```python
import nltk
nltk.download('punkt')
def tokenize(text):
  word=nltk.word_tokenize(text)
  return word
```

```python
data['tweet']=data['tweet'].apply(lambda x:tokenize(x))
```

```python
data
```

**Step4: Removal of stopwords**

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
  output=[i for i in text if i not in stopwords]
  return output
```

```python
data['tweet']=data['tweet'].apply(lambda x:remove_stopwords(x))
```

```python
data
```

**Step5:Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
def lemma(text):
  lemm_text=[lemmatizer.lemmatize(x) for x in text]
  return lemm_text
```

```python
data['tweet']=data['tweet'].apply(lambda x:lemma(x))
```

```python
data
```

```python
x=data.tweet
y=data.label
```

**TFIDF Vectorizer**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x= tfidf.fit_transform(x.astype('str'))
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**SMOTE-Oversampling Technique to nullify data imbalance**

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

**Model Evaluation**

```python
from sklearn.metrics import accuracy_score
y_predictions=rf.predict(x_test)
accuracy_score(y_predictions,y_test)
```
