**Preprocessing using Gensim**

```python
import gensim
```

```python
data['tweet']=data['tweet'].apply(lambda x:gensim.utils.simple_preprocess(x))
```

```python
data
```

**Removal of stopwords**

```python
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
  output=[i for i in text if i not in stopwords]
  return output
```

```python
data['tweet']=data['tweet'].apply(lambda x:remove_stopwords(x))
```

```python
data
```

**Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
def lemma(text):
  lemm_text=[lemmatizer.lemmatize(x) for x in text]
  return lemm_text
```

```python
data['tweet']=data['tweet'].apply(lambda x:lemma(x))
```

```python
data
```

**Separating dependant and independent variables**

```python
x=data.tweet
y=data.label
```

**Train test split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Training Word2vec**

```python
w2v_model=gensim.models.Word2Vec(x_train,min_count=1)
```

```python
w2v_model.wv['music']
```

```python
len(w2v_model.wv['music'])
```

```python
w2v_model.wv.most_similar('music')
```

```python
words=w2v_model.wv.index_to_key
```

```python
words
```

```python
len(words)
```

```python
import numpy as np
x_train_vec = np.array([np.array([w2v_model.wv[i] for i in ls if i in words])for ls in x_train])
x_test_vec = np.array([np.array([w2v_model.wv[i] for i in ls if i in words])for ls in x_test])
```

```python
x_train_vec_avg = []
for v in x_train_vec:
  if v.size:
    x_train_vec_avg.append(v.mean(axis=0))
  else:
    x_train_vec_avg.append(np.zeros(100,dtype = float))

x_test_vec_avg = []
for v in x_test_vec:
  if v.size:
    x_test_vec_avg.append(v.mean(axis=0))
  else:
    x_test_vec_avg.append(np.zeros(100,dtype = float))
```

**SMOTE-To nullify the effect of data imbalance**

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_vec_avg, y_train = smote.fit_resample(x_train_vec_avg, y_train)
```

**Model Training**

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier()
```

```python
rf_clf.fit(x_train_vec_avg,y_train)
```

```python
y_predictions=rf_clf.predict(x_test_vec_avg)
```

**Model Evaluation**

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predictions)
```
