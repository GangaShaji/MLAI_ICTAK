```python
def custom_encoder(data):
  data.replace(to_replace=['Negative emotion'],value=0,inplace=True)
  data.replace(to_replace=["I can't tell"],value=1,inplace=True)
  data.replace(to_replace=['No emotion toward brand or product'],value=2,inplace=True)
  data.replace(to_replace=['Positive emotion'],value=3,inplace=True)
  return data
```

```python
data['is_there_an_emotion_directed_at_a_brand_or_product']=custom_encoder(data['is_there_an_emotion_directed_at_a_brand_or_product'])
```

```python
data
```

```python
data.columns=['tweet','label']
```

```python
data
```

**Removing missing values**

```python
data.isna().sum()
```

```python
data=data.dropna()
```

```python
data.isna().sum()
```

**Tokenize** **the** **data**

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
```

```python
max_words = 5000  # Max vocabulary size
max_sequence_length = 30  # Max sequence length
embedding_dim = 50  # Embedding dimensions

tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(data['tweet'])
```

**Convert texts to sequences**

```python
data['tweet'] = tokenizer.texts_to_sequences(data['tweet'])
```

```python
data
```

**Separating dependent and independent variables**

```python
x=data.tweet
y=data.label
```

**Train test split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Padding sequences to make them equal in length**

```python
x_train=sequence.pad_sequences(x_train,maxlen=max_sequence_length)
x_test=sequence.pad_sequences(x_test,maxlen=max_sequence_length)
```

```python
x_train.shape
```

**SMOTE-to nullify data imbalance**

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
```

**One Hot Encoding target variable**

```python
y_train=pd.get_dummies(y_train)
y_test=pd.get_dummies(y_test)
```

# **Simple** **RNN**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
model = Sequential()
model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_sequence_length))
model.add(SimpleRNN(units=128, activation='relu',dropout=0.2))
model.add(Dense(units=4, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

# **LSTM**

```python
from tensorflow.keras.layers import LSTM
```

```python
model1 = Sequential()
model1.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_sequence_length))
model1.add(LSTM(units=128, activation='relu',dropout=0.2))
model1.add(Dense(units=4, activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model1.fit(x_train,y_train,batch_size=256,epochs=10)
```

```python
loss, accuracy = model1.evaluate(x_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

# **Convolution 1D**

```python
from tensorflow.keras.layers import Conv1D, Dropout, GlobalMaxPooling1D
```

```python
model3 = Sequential()
model3.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_sequence_length))
model3.add(Conv1D(64,kernel_size = 3, padding='same', activation = 'relu'))
model3.add(Dropout(0.2))
model3.add(GlobalMaxPooling1D())
model3.add(Dense(128, activation ='relu'))
model3.add(Dense(4, activation = 'softmax'))
```

```python
model3.summary()
```

```python
model3.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model3.fit(x_train,y_train,batch_size=256,epochs=10)
```

```python
loss, accuracy = model3.evaluate(x_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
